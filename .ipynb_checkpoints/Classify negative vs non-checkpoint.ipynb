```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import spacy
nlp = spacy.load('en')
%matplotlib inline
```

# DATA
Load in data from google's toxic comment dataset.
##### The dataset includes the following:
- id (str): unique id for each string
- comment_text (str): comment string
- labels (toxic, severe_toxic, obscene, threat, insult, identity_hate)

```python
train_data = pd.read_csv('data/toxic/train.csv')
test_data = pd.read_csv('data/toxic/test.csv')
labels = ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']
```

```python
# make sure there are no duplicate IDs
assert len(set(np.concatenate([train_data['id'].values, test_data['id'].values]))) == \
    len(train_data['id'].values) + len(test_data['id'].values), 'Duplicate IDs'
```

# Exploration of labels
First lets take a look at what type of problem this is. Right off the bat, it is not a binary classifier. The Kaggle page indicates that this is a multiclass multilabel classification problem.

- multiclass: type of classification problem where the number of classes exceeds two
- multilabel: type of classification problem where any given set of features can result in 0 to many classes

Consequently, the biggest thing we will need to look out far include imbalanced classes and correlations. If there are imbalanced classes, up/downsampling techniques may need to be used as well as class weights. Correlations mean that there may be benefits in not treating each class as a separate binary classifier

### Imbalanced class
there are almost 9x as many comments without labels as there are with labels. We can downsample the non-labeled comments; however, it may reduce the representativeness of non-labeled comments

```python
#label vs non_label
no_label_count = len(train_data[train_data[labels].sum(axis=1)==0])
label_count = len(train_data) - no_label_count
print('labeled', label_count)
print('non_labeled', no_label_count)
print(float(no_label_count)/float(label_count))
```

### Imbalanced class 2
those with labels are heavily skewed towards toxic comments

```python
#Label distribution
label_dist = train_data[labels].sum().reset_index().rename(columns={'index':'labels', 0:'counts'})
sns.barplot(x='labels', y='counts', data=label_dist)
```

### lets see how successful a prediction is for negative/non-negative comment

```python
train_data['NEGATIVE_COMMENT'] = train_data[labels].sum(axis=1) > 0
```

```python
print(train_data.groupby('NEGATIVE_COMMENT')['id'].count())
sns.countplot(x='NEGATIVE_COMMENT', data=train_data)
plt.show()
```

```python
#downsample some of the non-negatives
neg_non_df_train = train_data[train_data['NEGATIVE_COMMENT'] == 0].sample(16225, random_state=42)
neg_non_df_train = pd.concat([neg_non_df_train, train_data[train_data['NEGATIVE_COMMENT'] == 1]])
neg_non_df_train
```

```python
# test baseline naive bayes. In theory certain words should be able to indicate negative comments
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(neg_non_df_train['comment_text'], 
                                                    neg_non_df_train['NEGATIVE_COMMENT'], 
                                                    test_size=0.3)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB

tv = TfidfVectorizer()
nb = MultinomialNB()
pipe = Pipeline([('tfidf', tv), ('naive_bayes', nb)])
pipe.fit(X_train, y_train)
```

```python
from sklearn.metrics import f1_score, matthews_corrcoef, classification_report
preds = pipe.predict(X_test)
print(f1_score(y_test, preds))
print(matthews_corrcoef(y_test, preds))
print(classification_report(y_test, preds))
```

### baseline model
naive bayes is being used given an assumption that the style of writing for negative/non-negative comments is different. this also assumes that comments are non adversarial and do not use Bayesian poisoning. next easier steps is to remove stopwords, punctuation, and gridsearch

```python
neg_non_df_train['comment_text_spacy'] = neg_non_df_train['comment_text'].apply(lambda x: nlp(x))
```

```python
from spacy.lang.en.stop_words import STOP_WORDS
```

```python
neg_non_df_train['comment_text_sw'] = neg_non_df_train['comment_text_spacy'] \
                                    .apply(lambda x : ' '.join([tok.text for tok in x if tok.text not in STOP_WORDS]))
```

```python
neg_non_df_train['comment_text_sw']
```

```python
X_train, X_test, y_train, y_test = train_test_split(neg_non_df_train['comment_text_sw'], 
                                                    neg_non_df_train['NEGATIVE_COMMENT'], 
                                                    test_size=0.3)
```

```python
tv = TfidfVectorizer()
nb = MultinomialNB()
pipe = Pipeline([('tfidf', tv), ('naive_bayes', nb)])
pipe.fit(X_train, y_train)
```

```python
from sklearn.metrics import f1_score, matthews_corrcoef, classification_report
preds = pipe.predict(X_test)
print(f1_score(y_test, preds))
print(matthews_corrcoef(y_test, preds))
print(classification_report(y_test, preds))
```
